```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

Leitura da Imagem

```python
imagem = cv2.imread('imagem.jpg')
cv2_imshow(imagem)
```

Valores BGR do pixel na Origem da Imagem

```python
(b, g, r) = imagem[0, 0]
```

Altura x Largura da imagem em pixels

```python
altura, largura , _ = imagem.shape
```

Mostra os valores BGR nos cantos da Imagem

```python
px_sup_esq = imagem[0, 0]
print(f"Canto Superior Esquerdo (B, G, R): {px_sup_esq}")
px_sup_dir = imagem[0, largura - 1]
print(f"Canto Superior Direito (B, G, R): {px_sup_dir}")
px_inf_esq = imagem[altura - 1, 0]
print(f"Canto Inferior Esquerdo (B, G, R): {px_inf_esq}")
px_inf_dir = imagem[altura - 1, largura - 1]
print(f"Canto Inferior Direito (B, G, R): {px_inf_dir}")
```

Calcula a média de cor nos dois eixos da matriz ao mesmo tempo

```python
media_cor = np.mean(imagem, axis=(0, 1))
media_cor
```

Verifica qual dos valores de média é maior

```python
media_b, media_g, media_r = media_cor

if media_b > media_g and media_b > media_r:
    print("A imagem tem predominância de tons AZUIS.")
elif media_g > media_b and media_g > media_r:
    print("A imagem tem predominância de tons VERDES.")
else:
    print("A imagem tem predominância de tons VERMELHOS.")
```
