```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
imagem = cv2.imread('imagem.jpg')
```

```python
imagem_modificada = imagem.copy()
```

```python
linha, coluna = 100, 150
```

```python
nova_cor = [0,0,225]
```

```python
imagem_modificada[linha, coluna] = nova_cor
print(f"Pixel em ({linha}, {coluna}) alterado para {nova_cor}.")
```

```python
cv2.imwrite('imagem_modificada.jpg', imagem_modificada)
print("Imagem modificada foi salva como 'imagem_modificada.jpg'.")
```

```python
cv2_imshow(imagem_modificada)
```
