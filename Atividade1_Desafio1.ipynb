```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
imagem = cv2.imread('imagem.jpg')
cv2_imshow(imagem)
```

```python
(b, g, r) = imagem[0, 0]
```

```python
altura, largura , _ = imagem.shape
```

```python
px_sup_esq = imagem[0, 0]
print(f"Canto Superior Esquerdo (B, G, R): {px_sup_esq}")
px_sup_dir = imagem[0, largura - 1]
print(f"Canto Superior Direito (B, G, R): {px_sup_dir}")
px_inf_esq = imagem[altura - 1, 0]
print(f"Canto Inferior Esquerdo (B, G, R): {px_inf_esq}")
px_inf_dir = imagem[altura - 1, largura - 1]
print(f"Canto Inferior Direito (B, G, R): {px_inf_dir}")
```

```python
media_cor = np.mean(imagem, axis=(0, 1))
media_cor
```

```python
media_b, media_g, media_r = media_cor

if media_b > media_g and media_b > media_r:
    print("A imagem tem predominncia de tons AZUIS.")
elif media_g > media_b and media_g > media_r:
    print("A imagem tem predominncia de tons VERDES.")
else:
    print("A imagem tem predominncia de tons VERMELHOS.")
```
